```python
statuses = ['Current', 'Fully Paid', 'Late (16-30 days)', 'Late (31-120 days)',
            'Charged Off', 'Default', 'In Grace Period', 'Issued']
```

```python
sys.getsizeof(pd.Categorical(processed_first_chunk.PERIOD_END_LSTAT, statuses))/1000000
```

```python
def process(batch):
    batch = batch[columns].copy()
    for c in date_columns:
        # print('Processing column ', c)
        batch[c] = pd.to_datetime(batch[c])
    batch.PERIOD_END_LSTAT = pd.Categorical(batch.PERIOD_END_LSTAT, statuses)
    return batch
```

```python
processed_first_chunk = process(first_chunk)
```
